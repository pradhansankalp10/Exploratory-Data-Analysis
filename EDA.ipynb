```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/Processed_Animal_Dataset.csv")

# Display basic information about the dataset
print("Dataset Summary:")
print(df.info())  # Overview of data types and non-null counts
print("\nBasic Descriptive Statistics:")
print(df.describe(include='all'))  # Summary stats for all columns
print("\nMissing Values per Column:")
print(df.isnull().sum())  # Missing values in each column

# Display the first few rows to get a quick look at the data
print("\nDataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Select numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histogram for each numeric feature
for column in numeric_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('duration')
    plt.show()
```

```python
# Select categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Plot count plot for each categorical feature
for column in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=column, palette='viridis')
    plt.title(f'Count Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show() #Ensure that every plot is shown

# OR

#Add this line to the top of your code
%matplotlib inline

```

```python
# Plot box plot for each numeric feature
for column in numeric_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x=column, color='orange')
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
sns.pairplot(df[numeric_columns], diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle("Pair Plot of Numeric Features", y=1.02)
plt.show()
```

```python
from IPython import get_ipython
from IPython.display import display
# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/Processed_Animal_Dataset.csv")

# Display basic information about the dataset
print("Dataset Summary:")
print(df.info())  # Overview of data types and non-null counts
print("\nBasic Descriptive Statistics:")
print(df.describe(include='all'))  # Summary stats for all columns
print("\nMissing Values per Column:")
print(df.isnull().sum())  # Missing values in each column

# Display the first few rows to get a quick look at the data
print("\nDataset Head:")
print(df.head())

# %%
# Set style for seaborn
sns.set(style="whitegrid")

# Select numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histogram for each numeric feature
for column in numeric_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('duration')
    plt.show()

# %%
# Select categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Plot count plot for each categorical feature
for column in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=column, palette='viridis')
    plt.title(f'Count Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
# %%
# Ensure plots are displayed inline in Colab
%matplotlib inline

# Example: Plot for each combination of categorical and numeric columns
for cat_column in categorical_columns:
    for num_column in numeric_columns:
        plt.figure(figsize=(8, 5))
        sns.barplot(data=df, x=cat_column, y=num_column, ci='sd', palette='Blues')
        plt.title(f'{num_column} by {cat_column}')
        plt.xlabel(cat_column)
        plt.ylabel(f'Average {num_column}')
        plt.xticks(rotation=45)
        plt.show()
```
